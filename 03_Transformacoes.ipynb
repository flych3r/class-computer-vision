```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = 'files/bandeira_eua.jpg'
im = cv2.imread(filename)
```

```python
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
```

```python
width = im.shape[1]
height = im.shape[0]
```

```python
x_center = width / 2
y_center = height / 2
```

```python
M_rotation = cv2.getRotationMatrix2D((x_center, y_center), 45, 1)
print(M_rotation)
```

```python
im_rotated = cv2.warpAffine(im, M_rotation, (width, height))
plt.imshow(im_rotated)
```

```python
M_scaling = np.float32([[1, 0, 0], [0, 2, 0]])
print(M_scaling)
```

```python
im_scaled = cv2.warpAffine(im, M_scaling, (width, height))
plt.imshow(im_scaled)
```

```python
M_translated = np.float32([[1, 0, 0], [0, 1, 200]])
print(M_translated)
```

```python
im_translated = cv2.warpAffine(im, M_translated, (width, height))
plt.imshow(im_translated)
```
