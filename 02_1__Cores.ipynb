```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
eua_flag = cv2.imread('files/bandeira_eua.jpg', cv2.COLOR_BGR2RGB)
brazil_flag = cv2.imread('files/bandeira_brasil.jpeg', cv2.COLOR_BGR2RGB)
ck_img = cv2.imread('files/ck_fg2.jpg', cv2.COLOR_BGR2RGB)
```

```python
eua_flag = cv2.cvtColor(eua_flag, cv2.COLOR_BGR2RGB)
brazil_flag = cv2.cvtColor(brazil_flag, cv2.COLOR_BGR2RGB)
ck_img = cv2.cvtColor(ck_img, cv2.COLOR_BGR2RGB)
```

```python
_, bg_mask = cv2.threshold(ck_img[:,:,1], 254, 255,cv2.THRESH_BINARY_INV)
fg_mask = cv2.bitwise_not(bg_mask)
```

```python
bg_mask_3 = cv2.cvtColor(bg_mask, cv2.COLOR_GRAY2RGB)
plt.imshow(bg_mask, cmap='gray')
```

```python
fg_mask_3 = cv2.cvtColor(fg_mask, cv2.COLOR_GRAY2RGB)
plt.imshow(fg_mask, cmap='gray')
```

```python
ck_img.shape, bg_mask_3.shape
```

```python
bg_final = cv2.resize(eua_flag, (852, 480))
```

```python
img_bg_and = cv2.bitwise_and(ck_img, bg_mask_3)
plt.imshow(img_bg_and)
```

```python
img_fg_and = cv2.bitwise_and(bg_final, fg_mask_3)
plt.imshow(img_fg_and)
```

```python
cv2.imshow('Resultado', cv2.add(img_bg_and, img_fg_and))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
