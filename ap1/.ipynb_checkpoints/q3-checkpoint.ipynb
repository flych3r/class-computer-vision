```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_3 = cv2.imread('imagens/3.jpeg', 0)
plt.imshow(img_3, cmap='gray');
```

```python
img_3_fft = np.fft.fftshift(cv2.dft(np.float32(img_3), flags= cv2.DFT_COMPLEX_OUTPUT))
img_magnitude = np.log(cv2.magnitude(img_3_fft[:,:,0], img_3_fft[:,:,1]))
plt.imshow(img_magnitude, cmap='gray');
```

```python
l = img_3.shape[0]
c = img_3.shape[1]
```

```python
# filto da média para reduzir ruidos
mask = np.ones((3,3)) / 9.
```

```python
# mascara redimensionada (valores movidos para o centro)
filter_mask = np.zeros((l,c))
filter_mask[(l//2)-1:int(np.ceil(l/2))+1,(c//2)-1:int(np.ceil(c/2))+1] = mask
```

```python
# mascara no dominio da frequencia (fft e shift)
fft_mask = np.fft.fftshift(cv2.dft(np.float32(filter_mask), flags = cv2.DFT_COMPLEX_OUTPUT))
mask_abs = cv2.magnitude(fft_mask[:,:,0],fft_mask[:,:,1])
```

```python
# replica a magnitude da mascara para os dois canais
fft_mask[:,:,0] = mask_abs
fft_mask[:,:,1] = mask_abs
```

```python
# filtra no dominio da frequencia
fft_filtered = cv2.multiply(img_3_fft, fft_mask)
```

```python
img_magnitude_filtered = np.log(cv2.magnitude(fft_filtered[:,:,0], fft_filtered[:,:,1]))
plt.imshow(img_magnitude_filtered, cmap='gray');
```

```python
# inverte a FFT
img_back = cv2.idft(np.fft.ifftshift(fft_filtered))
img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])
```

```python
plt.imshow(img_back, cmap='gray');
```
