```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_4a = cv2.imread('imagens/4a.jpeg')
img_4a = cv2.cvtColor(img_4a, cv2.COLOR_BGR2RGB)
img_4a[np.where((img_4a == [255,255,255]).all(axis=2))] = [0, 0, 0]
plt.imshow(img_4a);
```

```python
img_4b = cv2.imread('imagens/4b.jpeg')
img_4b = cv2.cvtColor(img_4b, cv2.COLOR_BGR2RGB)
img_4b[np.where((img_4b == [255,255,255]).all(axis=2))] = [0, 0, 0]
plt.imshow(img_4b);
```

```python
img_4c = cv2.imread('imagens/4c.jpeg')
img_4c = cv2.cvtColor(img_4c, cv2.COLOR_BGR2RGB)
img_4c[np.where((img_4c == [255,255,255]).all(axis=2))] = [0, 0, 0]
plt.imshow(img_4c);
```

```python
img_4a.shape, img_4b.shape, img_4c.shape
```

```python
img_4c_border = cv2.copyMakeBorder(img_4c, 20, 20, 25, 25, cv2.BORDER_CONSTANT,value=[0,0,0])
img_4c_border = cv2.resize(img_4c_border, (246, 172))
plt.imshow(img_4c_border)
```

```python
green_yellow = cv2.add(img_4b, img_4c_border)
plt.imshow(green_yellow)
```

```python
img_4a_border = cv2.copyMakeBorder(img_4a, 60, 60, 130, 130, cv2.BORDER_CONSTANT,value=[0,0,0])
img_4a_border = cv2.resize(img_4a_border, (246, 172))
plt.imshow(img_4a_border)
```

```python
green_yellow_blue = cv2.add(green_yellow, img_4a_border)
plt.imshow(green_yellow_blue)
```
